# Adding 2 lines in same plot 

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(1,11)
y1 = 2*x
y2 = 4*x
plt.plot(x,y1, color='y', linestyle=":", linewidth=3)
plt.plot(x,y2, color='g', linestyle= '-.', linewidth=3)
plt.title('Adding 2 line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-label')
plt.grid(True)
```
