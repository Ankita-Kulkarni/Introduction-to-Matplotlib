```python
import matplotlib.pyplot as plt
fruits = ['Mango', 'Apple', 'Orange', 'Banana']
quantity = [90, 48, 12, 67]
plt.pie(quantity, labels=fruits)
plt.show()
```

# Changing aesthetics

```python
plt.pie(quantity, labels=fruits, autopct='%0.1f%%', colors=['yellow', 'grey', 'pink','cyan'])
plt.show()
```
