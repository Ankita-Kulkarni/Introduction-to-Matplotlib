1. Scatter plot is used to understand distribution between two numerical entities and these entities are represented in the form of data points.
2. Two list should have same no. of elements.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = [8,2,3,6]
plt.scatter(x,y, color='r')
```

```python
x = [10,20,30,40,60,70]
y = [8,2,3,6,5,4]
plt.scatter(x,y, c='orange', marker="*", s=300) #s=size
```

# Adding 2 marker in same plot

```python
x = [10,20,30,40,60,70]
y = [8,2,3,6,5,4]
z = [4,1,10,12,5,7] 
plt.scatter(x,y, c='orange', marker="*", s=300) 
plt.scatter(x,z, s=100)
```

# Adding subplot

```python
x = [10,20,30,40,60,70]
y = [8,2,3,6,5,4]
z = [4,1,10,12,5,7] 

plt.subplot(1,2,1)
plt.scatter(x,y, c='orange', marker="*", s=300) 

plt.subplot(1,2,2)
plt.scatter(x,z, s=100)
```

```python
x = [10,20,30,40,60,70]
y = [8,2,3,6,5,4]
z = [4,1,10,12,5,7] 

plt.subplot(2,1,1)
plt.scatter(x,y, c='orange', marker="*", s=300) 

plt.subplot(2,1,2)
plt.scatter(x,z, s=100)
```
