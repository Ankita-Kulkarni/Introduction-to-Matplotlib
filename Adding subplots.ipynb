# Adding subplots

# 1 row 2 cols

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x
plt.subplot(1,2,1)   # row=1, col=2, index=1
# At index=1 plot will be
plt.plot(x,y1, color='b', linestyle=":", linewidth=3)
plt.title('Adding subplots')
plt.grid(True)

plt.subplot(1,2,2)   # row=1, col=2, index=2
# At index=2 plot will be
plt.plot(x,y2, color='orange', linestyle= '-.', linewidth=3)

plt.title('Adding subplots')
plt.grid(True)
```

# 2 rows 1 col

```python
y1 = 2*x
y2 = 3*x
plt.subplot(2,1,1)   # row=2, col=1, index=1
# At index=1 plot will be
plt.plot(x,y1, color='b', linestyle=":", linewidth=3)
plt.title('Adding subplots')
plt.grid(True)

plt.subplot(2,1,2)   # row=2, col=1, index=2
# At index=2 plot will be
plt.plot(x,y2, color='orange', linestyle= '-.', linewidth=3)

plt.title('Adding subplots')
plt.grid(True)
```
