Similar to box plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7,8,9]
y = [2,3,4,2,7,6,3,4,8]
z = [1,4,6,2,6,2,1,2,9]

data = list([x,y,z])
plt.violinplot(data)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.title('Plot without showmedians')
plt.violinplot(data)

plt.subplot(1,2,2)
plt.title('Plot with showmedians')
plt.violinplot(data, showmedians=True)
plt.show()
```
